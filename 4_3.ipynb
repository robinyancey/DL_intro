# Data Preprocessing, Feature Engineering, and Feature Learning

## Introduction

In machine learning, the quality of the data significantly impacts the performance of the model. Data preprocessing, feature engineering, and feature learning are crucial steps to prepare data for training and improve model accuracy.

## Data Preprocessing for Neural Networks

Neural networks are sensitive to the scale and distribution of input data. Proper preprocessing ensures that the network can learn effectively and converge faster.

### Scaling Data

Neural networks often perform better when input features are scaled to a similar range. Common scaling techniques include:

#### Min-Max Scaling

Scales features to a specified range, typically [0, 1].

Formula:

$$ X_{scaled} = \frac{X - X_{min}}{X_{max} - X_{min}} $$

#### Standardization (Z-score normalization)

Scales features to have zero mean and unit variance.

Formula:

$$ X_{scaled} = \frac{X - \mu}{\sigma} $$

Where $\mu$ is the mean and $\sigma$ is the standard deviation.

### Handling Missing Values

Missing values can introduce bias and affect model performance. Common strategies include:

*   **Imputation:** Replacing missing values with a statistical measure (mean, median, mode) or using more advanced techniques like K-Nearest Neighbors (KNN) imputation.
*   **Dropping:** Removing rows or columns with missing values (use with caution to avoid losing valuable data).

### Encoding Categorical Data

Neural networks typically require numerical input. Categorical features need to be converted into a numerical representation.

#### One-Hot Encoding

Creates binary columns for each category.

Example: If a feature 'Color' has categories 'Red', 'Green', 'Blue', one-hot encoding creates columns 'Color_Red', 'Color_Green', 'Color_Blue' with binary values.

#### Label Encoding

Assigns a unique integer to each category.

Example: 'Red' becomes 0, 'Green' becomes 1, 'Blue' becomes 2. This can imply an ordinal relationship, which might not be appropriate for all categorical data.

### Code Examples: Data Preprocessing

Let's demonstrate some preprocessing techniques using Python's scikit-learn library.

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
import pandas as pd

# Load the Titanic dataset
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
print("Original DataFrame (Titanic Dataset):")
display(df.head())

# Select relevant features for preprocessing (handling missing values, scaling, encoding)
# We will use 'Age', 'Fare', 'Embarked', 'Sex' as examples
df_subset = df[['Age', 'Fare', 'Embarked', 'Sex']].copy()

# Handling missing values in 'Age' using mean imputation
imputer_age = SimpleImputer(missing_values=np.nan, strategy='mean')
df_subset['Age'] = imputer_age.fit_transform(df_subset[['Age']])
print("\nDataFrame after mean imputation for 'Age':")
display(df_subset.head())

# Handling missing values in 'Embarked' using most frequent imputation
imputer_embarked = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
df_subset['Embarked'] = imputer_embarked.fit_transform(df_subset[['Embarked']]).ravel()
print("\nDataFrame after most frequent imputation for 'Embarked':")
display(df_subset.head())

# Min-Max Scaling for 'Age' and 'Fare'
scaler_minmax = MinMaxScaler()
df_subset[['Age', 'Fare']] = scaler_minmax.fit_transform(df_subset[['Age', 'Fare']])
print("\nDataFrame after Min-Max Scaling for 'Age' and 'Fare':")
display(df_subset.head())

# Standardization for 'Age' and 'Fare'
# Note: Applying both Min-Max and Standardization in sequence is just for demonstration
# In a real scenario, you would typically choose one based on the model and data distribution
scaler_standard = StandardScaler()
df_subset[['Age', 'Fare']] = scaler_standard.fit_transform(df_subset[['Age', 'Fare']])
print("\nDataFrame after Standardization for 'Age' and 'Fare':")
display(df_subset.head())

# One-Hot Encoding for 'Embarked' and 'Sex'
encoder = OneHotEncoder(sparse_output=False)
encoded_categories = encoder.fit_transform(df_subset[['Embarked', 'Sex']])
encoded_df = pd.DataFrame(encoded_categories, columns=encoder.get_feature_names_out(['Embarked', 'Sex']))

# Concatenate the original subset (without the original categorical columns) and the encoded features
df_processed = df_subset.drop(['Embarked', 'Sex'], axis=1)
df_processed = pd.concat([df_processed, encoded_df], axis=1)

print("\nDataFrame after One-Hot Encoding for 'Embarked' and 'Sex':")
display(df_processed.head())

# Store the processed dataframe for the next steps
processed_df = df_processed.copy()
```

## Feature Engineering

Feature engineering is the process of creating new features from existing ones to improve model performance. This often involves domain knowledge and creativity.

### Creating Polynomial Features

Generating polynomial combinations of existing features can capture non-linear relationships.

Example: From features $X_1$ and $X_2$, create $X_1^2$, $X_2^2$, $X_1X_2$.

### Interaction Features

Combining features to represent interactions between them.

Example: Multiplying two features to see if their combined effect is significant.

### Extracting Date and Time Features

From a datetime column, extract features like year, month, day, day of the week, hour, etc.

### Binning Numerical Features

Grouping numerical values into bins (intervals) to create categorical features. This can help capture non-linear relationships or reduce the impact of outliers.

### Code Examples: Feature Engineering

Let's look at some examples of feature engineering.

```python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures

# Use the processed dataframe from the previous step
df = processed_df.copy()
print("DataFrame before Feature Engineering (Processed Titanic Data):")
display(df.head())

# Creating Polynomial Features (degree 2) for 'Age' and 'Fare'
# Note: This is for demonstration; applying polynomial features to scaled/standardized data might require careful consideration
poly = PolynomialFeatures(degree=2, include_bias=False)
# Select only numerical columns for polynomial features
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()
poly_features = poly.fit_transform(df[numerical_cols])
poly_df = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(numerical_cols))

# Drop the original numerical columns and concatenate the polynomial features
df_poly = df.drop(numerical_cols, axis=1)
df_poly = pd.concat([df_poly, poly_df], axis=1)
print("\nDataFrame after Polynomial Features (for numerical columns):")
display(df_poly.head())

# Creating Interaction Feature (example: Age * Fare)
# Use the original processed dataframe to create the interaction feature
df['Age_Fare_Interaction'] = df['Age'] * df['Fare']
print("\nDataFrame after Interaction Feature (Age * Fare):")
display(df.head())

# Extracting Date and Time Features
# The Titanic dataset does not have a datetime column, so we will skip this section
print("\nSkipping Date and Time Feature Extraction as the Titanic dataset does not have a datetime column.")

# Binning Numerical Features (example with Age)
# Create bins for Age (e.g., Child, Young, Adult, Senior)
# Use the Age column before standardization/scaling for meaningful binning
original_age = df_subset['Age'] # Use the age after mean imputation from the previous cell
bins = [0, 12, 18, 60, 120]
labels = ['Child', 'Teenager', 'Adult', 'Senior']
df['Age_Binned'] = pd.cut(original_age, bins=bins, labels=labels, right=True)
print("\nDataFrame after Binning Age:")
display(df.head())

# Store the feature-engineered dataframe for the next steps (optional, depending on subsequent examples)
feature_engineered_df = df.copy()
```

## Feature Learning

Feature learning (or representation learning) refers to a set of techniques that allow a system to automatically discover the representations needed for feature detection or classification from raw data. This is a key advantage of deep learning, where the network learns hierarchical representations of the data through its layers.

### Deep Learning and Feature Learning

Neural networks, especially deep neural networks, can learn complex features directly from the data during the training process. Lower layers often learn basic features (e.g., edges in images), while higher layers learn more abstract and complex features.

### Examples of Feature Learning Techniques

*   **Autoencoders:** Neural networks trained to reconstruct their input. The hidden layer learns a compressed representation (features) of the input data.
*   **Convolutional Neural Networks (CNNs):** Primarily used for image data, CNNs learn hierarchical features through convolutional and pooling layers.
*   **Recurrent Neural Networks (RNNs):** Used for sequential data, RNNs learn features that capture temporal dependencies.

### Code Examples: Feature Learning (Conceptual)

Implementing full feature learning models like Autoencoders or CNNs requires building and training neural networks, which is beyond a simple code snippet. However, we can illustrate the concept with a basic example of how a neural network layer can be seen as learning a new representation.

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.models import Model
import numpy as np
import pandas as pd

# Use the feature-engineered dataframe from the previous step as conceptual input
# In a real scenario, this would be your training data (features)
input_data = feature_engineered_df.select_dtypes(include=np.number).values # Select numerical features as input

print("Conceptual Input Data for Feature Learning (Numerical features from Titanic Dataset):")
display(pd.DataFrame(input_data).head()) # Display as DataFrame for better readability

# Define a simple neural network with one hidden layer
# This is a conceptual example to show how a layer learns representations
input_layer = Input(shape=(input_data.shape[1],)) # Input shape based on the number of features
hidden_layer = Dense(3, activation='relu')(input_layer) # This layer learns a new representation
output_layer = Dense(1)(hidden_layer) # Conceptual output layer

model = Model(inputs=input_layer, outputs=output_layer)

# We can access the learned weights of the hidden layer after training
# For demonstration, let's just show the structure and a dummy prediction
model.compile(optimizer='adam', loss='mse')

# Dummy training (in a real scenario, you would train on a dataset with labels)
# model.fit(input_data, target_data, epochs=...)

print("\nModel Summary (showing the hidden layer as learning a representation):")
model.summary()

# Conceptual: The output of the hidden layer for the input data is the learned feature representation
# Note: The actual learned representation depends on the training process and weights
# For demonstration, let's just show the output of the hidden layer with untrained weights
hidden_layer_model = Model(inputs=input_layer, outputs=hidden_layer)
learned_features = hidden_layer_model.predict(input_data)

print("\nConceptual Learned Features (output of the hidden layer with untrained weights):")
display(pd.DataFrame(learned_features).head()) # Display as DataFrame
```
