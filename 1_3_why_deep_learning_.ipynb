## 1. Hardware

*   **Increased computational power:** The advent of powerful GPUs (Graphics Processing Units) and TPUs (Tensor Processing Units) has significantly accelerated the training of deep neural networks.
*   **Parallel processing:** These specialized hardware units are designed for parallel processing, which is crucial for the computationally intensive matrix operations involved in deep learning.
*   **Falling hardware costs:** The cost of powerful hardware has decreased over time, making deep learning more accessible to researchers and developers.

### Code Example: Illustrating the speedup with matrix multiplication

Let's illustrate the difference in speed between matrix multiplication on a CPU and a GPU (or a simulated one).

```python
import numpy as np
import time

# Create large matrices
matrix_size = 1000
matrix_a = np.random.rand(matrix_size, matrix_size)
matrix_b = np.random.rand(matrix_size, matrix_size)

# Simulate CPU multiplication
start_time_cpu = time.time()
result_cpu = np.dot(matrix_a, matrix_b)
end_time_cpu = time.time()
cpu_time = end_time_cpu - start_time_cpu
print(f"CPU time: {cpu_time:.4f} seconds")

# Note: To truly show GPU speedup, you would need a GPU and a library like TensorFlow or PyTorch.
# This is a conceptual example.

# Simulate faster GPU multiplication (conceptually)
# In a real scenario, GPU multiplication would be significantly faster for large matrices
gpu_time = cpu_time * 0.01 # Illustrative speedup
print(f"Simulated GPU time: {gpu_time:.4f} seconds")
```

### Code Example: Visualizing Data Size Impact

Let's visualize how the performance of a simple model might improve with increasing data size.

```python
import matplotlib.pyplot as plt
import numpy as np

# Simulate model performance vs. data size
data_sizes = np.array([100, 500, 1000, 5000, 10000])
# Simulate performance (e.g., accuracy) improving with data size
performance = 1 - 1 / np.sqrt(data_sizes) * 2  # Example relationship

plt.figure(figsize=(8, 6))
plt.plot(data_sizes, performance, marker='o', linestyle='-')
plt.xlabel("Data Size")
plt.ylabel("Model Performance (e.g., Accuracy)")
plt.title("Simulated Model Performance vs. Data Size")
plt.grid(True)
plt.show()
```

### Code Example: Illustrating Activation Functions

Let's visualize some common activation functions used in deep learning. Activation functions introduce non-linearity into the network, enabling it to learn complex patterns.

```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

x = np.linspace(-5, 5, 100)

plt.figure(figsize=(10, 6))
plt.plot(x, sigmoid(x), label='Sigmoid')
plt.plot(x, relu(x), label='ReLU')
plt.plot(x, tanh(x), label='Tanh')
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Common Activation Functions")
plt.legend()
plt.grid(True)
plt.show()
```

### Code Example: Illustrating a simple neural network structure

Let's visualize a simple neural network structure.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add nodes for layers
G.add_node("Input Layer", pos=(0, 1.5))
G.add_node("Hidden Layer 1", pos=(1, 2))
G.add_node("Hidden Layer 2", pos=(1, 1))
G.add_node("Output Layer", pos=(2, 1.5))

# Add edges (connections)
G.add_edges_from([("Input Layer", "Hidden Layer 1"), ("Input Layer", "Hidden Layer 2")])
G.add_edges_from([("Hidden Layer 1", "Output Layer"), ("Hidden Layer 2", "Output Layer")])


# Get positions for nodes
pos = nx.get_node_attributes(G, 'pos')

# Draw the graph
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='skyblue', font_size=10, font_weight='bold', arrowsize=20)
plt.title("Simple Neural Network Structure")
plt.axis('off')
plt.show()
```
